# EDA AND Feaure Engineering Flight Price Prediction


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head(2)
```

```python
type(df['Date_of_Journey'].str.split('/').str[0][0])
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head(2)
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey' , axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.head(10)
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Hours'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Arrival_Hours'] = df['Arrival_Hours'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python
df.info()
```

```python
df.head(5)
```

```python
df['Dept_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dept_min'] = df['Dep_Time'].str.split(':').str[1]

df['Dept_hour'] = df['Dept_hour'].astype(int)
df['Dept_min'] = df['Dept_min'].astype(int)

df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.drop('Route' , axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration'].str.split(' ').str[1].str.split('h').str[0]

```

```python
df['Duration'] = df['Duration'].astype(int)

df.drop('Duration',axis=1,inplace=True)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop' : 1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1 })
```

```python
df['Total_Stops']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder() 
```

```python
encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```
